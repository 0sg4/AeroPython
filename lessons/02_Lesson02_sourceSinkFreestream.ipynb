We start by importing the libraries that we will be using in this notebook.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
from IPython.core.display import clear_output
```

To visualize the streamlines, a mesh grid has to be created.

```python
N = 200                           # Number of points in each direction
xStart,xEnd = -4.0,4.0            # x-direction boundaries
yStart,yEnd = -2.0,2.0            # y-direction boundaries
x = np.linspace(xStart,xEnd,N)    # x 1D-array
y = np.linspace(yStart,yEnd,N)    # y 1D-array
X,Y = np.meshgrid(x,y)            # generation of the mesh grid
```

## Source in a uniform flow

In the previous notebook, we computed the velocity field induced by a source ans a sink. The objective here is to superimpose a source on a uniform flow.

The streamlines of a freestream (speed $U_\infty$, angle of attack $\alpha$) are given by $\psi = \text{Const}$ where:

$$\psi_\text{freestream} = U_\infty \left(y\cos\alpha - x\sin\alpha\right)$$

```python
Uinf = 1.0        # freestream speed
alpha = 0.0       # angle of attack (in degrees)

# computing the velocity components on the mesh grid
uFreestream = Uinf*cos(alpha*pi/180)*np.ones((N,N),dtype=float)
vFreestream = Uinf*sin(alpha*pi/180)*np.ones((N,N),dtype=float)

# computing the stream-function on the mesh grid
psiFreestream = + Uinf*cos(alpha*pi/180)*Y\
                - Uinf*sin(alpha*pi/180)*X
```

We recall that the stream function of a source flow located at $\left(x_\text{source},y_\text{source}\right)$ is:

$$\psi_\text{source} = \frac{\sigma}{2\pi} \arctan \left(\frac{y-y_\text{source}}{x-x_\text{source}}\right)$$

In the previous notebook, we also saw that the stream function and velocity field of a source and a sink were exactly the same except the sign of the strength. Therefore, instead of rewriting the same code for the source and for the sink, we should embed the velocity calculation and the stream-function calculation inside functions.

To do that, we define a function called `getVelocity` to compute the velocity components (`u`,`v`) on a mesh (`X`,`Y`), given the `strength` of the singularity and its location (`xs`,`ys`).

```python
# function to compute the velocity field of a source/sink
def getVelocity(strength,xs,ys,X,Y):
    u = strength/(2*pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v = strength/(2*pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
    return u,v
```

In a similar manner, we define another function, `getStreamFunction`, to compute the stream-function of the singularity (source or sink) on the mesh grid.

```python
# function to compute the stream-function of a source/sink
def getStreamFunction(strength,xs,ys,X,Y):
    psi = strength/(2*pi)*np.arctan2((Y-ys),(X-xs))
    return psi
```

Now, we can compute the velocity field and the stream-function of a source flow using those two functions. In the future (in this notebook), adding an other singularity to the flow such as a sink will be easy as pie!

```python
strengthSource = 5.0         # strength of the source
xSource,ySource = -1.0,0.0   # location of the source

# computing the velocity components
uSource,vSource = getVelocity(strengthSource,xSource,ySource,X,Y)

# computing the stream-function
psiSource = getStreamFunction(strengthSource,xSource,ySource,X,Y)
```

Since both stream-functions are solutions of Laplace's equation ($\nabla^2 \psi=0$), the sum also satisfies Laplace's equation. The streamlines of the combination of a freestream and a source flow are obtained from:

$$\psi = \psi_{freestream}+\psi_{source} = U_\infty \left(y\cos\alpha - x\sin\alpha\right) + \frac{\sigma}{2\pi} \arctan \left(\frac{y-y_\text{source}}{x-x_\text{source}}\right)$$

And since differentiation is linear, the velocity field induced by the new flow pattern is simply the sum of the freestream velocity field and the source velocity field:

$$u = u_\text{freestream} + u_\text{source}$$
$$v = v_\text{freestream} + v_\text{source}$$

The stagnation point in the flow is obtained by solving the equations:

$$u = 0 \quad \text{,} \quad v = 0$$

which leads to point $x_\text{stagnation},y_\text{stagnation}$:

$$x_\text{stagnation} = x_\text{source} + \frac{\sigma}{2\pi U_\infty}\cos \left(\alpha+\pi\right)$$

$$y_\text{stagnation} = y_\text{source} + \frac{\sigma}{2\pi U_\infty}\sin \left(\alpha+\pi\right)$$

The streamline containing the stagnation is called the *dividing streamline*. It seperates the fluid coming from the freestream to the fluid radiating form the source flow.

```python
# superposition of the source on the freestream
u = uFreestream + uSource
v = vFreestream + vSource
psi = psiFreestream + psiSource

# plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xSource,ySource,c='#CD2305',s=80,marker='o')
plt.contour(X,Y,psi,\
            levels=[strengthSource/2,-strengthSource/2],\
            colors='#CD2305',linewidths=2,linestyles='solid')

xStagnation = xSource + strengthSource/(2*pi*Uinf)*cos(pi+alpha)
yStagnation = ySource + strengthSource/(2*pi*Uinf)*sin(pi+alpha)
plt.scatter(xStagnation,yStagnation,c='b',s=80,marker='o')
```

## Source-sink pair in a uniform flow

We would like to add a sink to our flow pattern. Well, now we can enjoy the usefulness of our two functions `getVelocity` and `getStreamfunction`, to create a sink.

```python
strengthSink = -5.0     # strength of the sink
xSink,ySink = 1.0,0.0   # location of the sink

# computing the velocity field on the mesh grid
uSink,vSink = getVelocity(strengthSink,xSink,ySink,X,Y)

# computing the stream-function on the grid mesh
psiSink = getStreamFunction(strengthSink,xSink,ySink,X,Y)
```

The superimposition of the freestream, the source and the sink can now be computed (just simple additions).

```python
# superposition of a source and a sink on the freestream
u = uFreestream + uSource + uSink
v = vFreestream + vSource + vSink
psi = psiFreestream + psiSource + psiSink

# plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter([xSource,xSink],[ySource,ySink],c='#CD2305',s=80,marker='o')
plt.contour(X,Y,psi,levels=[0.0],colors='#CD2305',linewidths=2,linestyles='solid')
```

A very useful measurement of a flow around a body is the *coefficient of pressure* $C_p$. To evaluate the pressure coefficient, we apply *Bernoulli's equation* for an imcompressible flow:

$$p_\infty + \frac{1}{2}\rho V_\infty^2 = p + \frac{1}{2}\rho V^2$$

We define the pressure coefficient:

$$C_p = \frac{p-p_\infty}{\frac{1}{2}\rho V_\infty^2}$$

In an incompressible flow, $C_p=1$ at a stagnation point.

```python
# computing the pressure coefficient
Cp = 1.0-(u**2+v**2)/Uinf**2

# plotting
size = 10
plt.figure(num=0,figsize=(1.1*size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
contf = plt.contourf(X,Y,Cp,levels=np.linspace(-2.0,1.0,100),extend='both')
cbar = plt.colorbar(contf)
cbar.set_label(r'$C_p$',fontsize=16)
cbar.set_ticks([-2.0,-1.0,0.0,1.0])
plt.scatter([xSource,xSink],[ySource,ySink],c='#CD2305',s=80,marker='o')
plt.contour(X,Y,psi,\
            levels=[0.0],\
            colors='#CD2305',linewidths=2,linestyles='solid')
```
