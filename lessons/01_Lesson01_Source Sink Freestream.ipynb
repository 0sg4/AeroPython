# Free-stream, source and sink

We start by importing libraries to help us out:

* `numpy` is a scientific library to create and manage multi-dimensional arrays and matrices.
* `matplotlib` is a 2D plotting library we are going to use to visualize our results.
* The `math` module provides access to the mathematical functions defined by the C standard.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
from IPython.core.display import clear_output
```

First, we would like to define an evenly spaced Cartesian grid of points within a spatial domain that is 4 units of length wide in the $x$-direction and 2 units of length wide in the $y$-direction, i.e. $x,y\in\left[-2,2\right],\left[-1,1\right]$. We define `N` the number of points  we want in the two directions, as well as the computational boundaries `xStart`,`xEnd`,`yStart` and `yEnd`. Then, we use the `numpy` function `linspace` to create two 1D arrays that contain the evenly spaced values of $x$ and $y$. The last line calls an other function of numpy `meshgrid` to generate a grid containing the coordinates of points where the numerical solution will be calculated.

```python
N = 50                            # Number of points in each direction
xStart,xEnd = -2.0,2.0            # x-direction boundaries
yStart,yEnd = -1.0,1.0            # y-direction boundaries
x = np.linspace(xStart,xEnd,N)    # x 1D-array
y = np.linspace(yStart,yEnd,N)    # y 1D-array
print 'x = ',x
print 'y = ',y
X,Y = np.meshgrid(x,y)            # generation of the mesh grid
```

Now the mesh grid has been generated, it is time to visualize it with the module `pyplot` from the library `matplotlib` using the alias `plt` and the function `scatter`.

```python
# plotting
size = 8
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.scatter(X,Y,s=8,c='b',marker='o')
```

## $\bullet$ Freestream flow

```python
Uinf,Vinf = 1.0,0.0               # freestream velocity magnitudes

# Cartesian velocity components
uF = Uinf*np.ones((N,N),dtype=float)
vF = Vinf*np.ones((N,N),dtype=float)

# plotting
size = 8
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uF,vF,color='b',density=0.5,linewidth=2,arrowsize=2,arrowstyle='->')
```

## $\bullet$ Source flow

Let consider a two-dimensional incompressible flow where all the streamlines emanate from a single point as straight lines and where the radial velocity is inversionally proportional to the distance from the point and the tangential velocity is zero. Such a flow is called a source.

$$u_r\left(r,\theta\right) = \frac{\sigma_\text{source}}{2\pi r} \quad \text{,} \quad u_\theta\left(r,\theta\right)=0$$
where $\sigma_\text{source}$ denotes the source strength and $\left(r,\theta\right)$ are the polar coordinates with the origin being at the source point.

In cylindrical coordinates, the stream function $\psi$ is obtained as follow:

$$\frac{1}{r}\frac{\partial\psi}{\partial\theta} = u_r \quad \text{,} \quad -\frac{\partial\psi}{\partial r} = u_\theta$$

leading to

$$\psi = \frac{\sigma_\text{source}}{2\pi}\theta + \text{constant}$$

In practical problems, we are more interested in the velocity components that are obtained by differenciation of the stream function, so that the constant can be dropped.

The source flow solution satisfies Laplace's equation $\nabla^2 \psi = 0$, so its simple solution could be used to build a more complex flow.

In Cartesian coordinates, the velocity field $\left(u,v\right)$ at position $\left(x,y\right)$ induced by the source $\sigma_\text{source}$ located at `xSource`,`ySource` are given by:

$$u = \frac{\partial\psi}{\partial y} = \frac{\sigma_\text{source}}{2\pi}\frac{x-\text{xSource}}{\sqrt{\left(x-\text{xSource}\right)^2+\left(y-\text{ySource}\right)^2}}$$

and

$$v = -\frac{\partial\psi}{\partial x} = \frac{\sigma_\text{source}}{2\pi}\frac{y-\text{ySource}}{\sqrt{\left(x-\text{xSource}\right)^2+\left(y-\text{ySource}\right)^2}}$$

```python
source = 5.0                          # source strength
xSource,ySource = -0.5,0.0            # location of the source

# Cartesian velocity components
uS = source/(2*pi)*(X-xSource)/np.sqrt((X-xSource)**2+(Y-ySource)**2)
vS = source/(2*pi)*(Y-ySource)/np.sqrt((X-xSource)**2+(Y-ySource)**2)

# plotting
size = 8
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uS,vS,density=1.0,linewidth=2,arrowsize=2,arrowstyle='->')
plt.scatter(xSource,ySource,s=40,c='r',marker='o')
```

## $\bullet$ Sink flow

In the previous paragraph, the strength $\sigma$ of the singularity was chosen to be positive. A source with a negative strength is called a sink. Instead of emanating from a single point, the straight streamlines are converging to a single point.

The velocity field induced by a sink is similar the one induced by a source, except that the singularity strength is negative ($\sigma_\text{sink} < 0$). And the Python code requires very few modifications.

```python
sink = -5.0                           # sink strength
xSink,ySink = 0.5,0.0                 # location of the sink

# Cartesian velocity components
uSi = sink/(2*pi)*(X-xSink)/np.sqrt((X-xSink)**2+(Y-ySink)**2)
vSi = sink/(2*pi)*(Y-ySink)/np.sqrt((X-xSink)**2+(Y-ySink)**2)

# plotting
size = 8
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uSi,vSi,density=1.0,linewidth=2,arrowsize=2,arrowstyle='->')
plt.scatter(xSink,ySink,s=40,c='r',marker='o')
```

## $\bullet$ Source superimposed on freestream

What is happening if we decide to superimposed a source on the freestream flow?

Since Laplace's is a linear differential equation, the stream function induced by the superimposition is just the sum of the stream function of a freestream and the one of the source, that is to say:

$$\psi = U_\infty r\sin\theta - V_\infty r\cos\theta + \frac{\sigma_\text{source}}{2\pi}\theta$$

And the resulting velocity field is simply the sum of the two velocity fields. Therefore, to visualize the streamlines induced, we just add `uS` to `uF`.

The velocity field is given by:

$$u = U_\infty + \frac{\sigma_\text{source}}{2\pi}\frac{x-\text{xSource}}{\sqrt{\left(x-\text{xSource}\right)^2+\left(y-\text{ySource}\right)^2}} + \frac{\sigma_\text{sink}}{2\pi}\frac{x-\text{xSink}}{\sqrt{\left(x-\text{xSink}\right)^2+\left(y-\text{ySink}\right)^2}}$$

$$v = V_\infty + \frac{\sigma_\text{source}}{2\pi}\frac{y-\text{ySource}}{\sqrt{\left(x-\text{xSource}\right)^2+\left(y-\text{ySource}\right)^2}} + \frac{\sigma_\text{sink}}{2\pi}\frac{y-\text{ySink}}{\sqrt{\left(x-\text{xSink}\right)^2+\left(y-\text{ySink}\right)^2}}$$

```python
u = uF + uS
v = vF + vS

size = 8
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,density=1.0,linewidth=2,arrowsize=2,arrowstyle='->')
plt.scatter(xSource,ySource,s=40,c='r',marker='o')
```

## $\bullet$ Adding a sink

Now, we are going to add a new singularity to the flow, a sink. The velocity field induced is the addition of the freestream velocity (`uF`,`vF`), the source velocity (`uS`,`vS`) and the sink velocity (`uSi`,`vSi`).

The stream function is given by:

$$\psi = U_\infty r\sin\theta - V_\infty r\cos\theta + \frac{\sigma_{\text{source}}}{2\pi}\theta \frac{\sigma_{\text{sink}}}{2\pi}\theta$$

and the velocity field is:

$$u = U_\infty + \frac{\sigma_\text{source}}{2\pi}\frac{x-\text{xSource}}{\sqrt{\left(x-\text{xSource}\right)^2+\left(y-\text{ySource}\right)^2}} + \frac{\sigma_\text{sink}}{2\pi}\frac{x-\text{xSink}}{\sqrt{\left(x-\text{xSink}\right)^2+\left(y-\text{ySink}\right)^2}}$$

$$v = V_\infty + \frac{\sigma_\text{source}}{2\pi}\frac{y-\text{ySource}}{\sqrt{\left(x-\text{xSource}\right)^2+\left(y-\text{ySource}\right)^2}} + \frac{\sigma_\text{sink}}{2\pi}\frac{y-\text{ySink}}{\sqrt{\left(x-\text{xSink}\right)^2+\left(y-\text{ySink}\right)^2}}$$


```python
u += uSi
v += vSi

size = 8
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,density=1.0,linewidth=2,arrowsize=2,arrowstyle='->')
plt.scatter(xSource,ySource,s=40,c='r',marker='o')
plt.scatter(xSink,ySink,s=40,c='r',marker='o')
```
