# Doublet

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
from IPython.core.display import clear_output
```

In the previous notebook, we saw that a source-sink pair in a uniform flow can be used  to represent the streamlines around a particular shape, named a 2Rankine oval. Let study the case where the distance of the source-sink pair approaches zero.

We consider a source of strength $\sigma$ at $\left(-\frac{l}{2},0\right)$ and a sink of opposite strength located at $\left(\frac{l}{2},0\right)$.

```python
from IPython.display import Image
Image(filename='../resources/doubletSketch1.png')
```

The stream-function induced by the source-sink pair at point $\text{P}\left(x,y\right)$ is

$$\psi\left(x,y\right) = \frac{\sigma}{2\pi}\left(\theta_1-\theta_2\right) = -\frac{\sigma}{2\pi}\Delta\theta$$

Let the distance $l$ between the two singularities approach zero while the strength magnitude is increasing so that the product $\sigma l$ remains constant. Such a flow pattern is called a *doublet* and we define the strength of the new singularity by $\kappa = \sigma l$.

The stream-function induced by a doublet at point $\text{P}\left(x,y\right)$ is given by

$$\psi\left(x,y\right) = \lim \limits_{l \to 0} \left(-\frac{\sigma}{2\pi}d\theta\right) \quad \text{and} \quad \sigma l = \text{constant}$$

```python
from IPython.display import Image
Image(filename='../resources/doubletSketch2.png')
```

Considering the case where $d\theta$ is infinitesimal, we deduce from the figure above that

$$a = l\sin\theta$$

$$b = r-l\cos\theta$$

$$d\theta = \frac{a}{b} = \frac{l\sin\theta}{r-l\cos\theta}$$

so that the stream-function becomes

$$\psi\left(x,y\right) = \lim \limits_{l \to 0} \left(-\frac{\sigma l}{2\pi}\frac{\sin\theta}{r-l\cos\theta}\right) \quad \text{and} \quad \sigma l = \text{constant}$$

i.e.

$$\psi\left(x,y\right) = -\frac{\kappa}{2\pi}\frac{\sin\theta}{r}$$

In Cartesian coordinates, a doublet located at the origin is given by

$$\psi\left(x,y\right) = -\frac{\kappa}{2\pi}\frac{y}{x^2+y^2}$$

from which we can derive the velocity components

$$u\left(x,y\right) = \frac{\partial\psi}{\partial y} = -\frac{\kappa}{2\pi}\frac{x^2-y^2}{\left(x^2+y^2\right)^2}$$

$$v\left(x,y\right) = -\frac{\partial\psi}{\partial x} = -\frac{\kappa}{2\pi}\frac{2xy}{\left(x^2+y^2\right)^2}$$

Now we have done the math, it is time to code and visualize what the streamlines look like. We start by creating a mesh grid.

```python
N = 50                            # Number of points in each direction
xStart,xEnd = -2.0,2.0            # x-direction boundaries
yStart,yEnd = -1.0,1.0            # y-direction boundaries
x = np.linspace(xStart,xEnd,N)    # x 1D-array
y = np.linspace(yStart,yEnd,N)    # y 1D-array
X,Y = np.meshgrid(x,y)            # generation of the mesh grid
```

We consider a doublet of strength $\kappa=1.0$ located at the origin.

```python
kappa = 1.0                    # strength of the doublet
xDoublet,yDoublet = 0.0,0.0    # location of the doublet
```

As seen in the previous notebook, we play smart by defining functions to calculate the stream-function and the velocity components that could be re-used if we decide to insert more than one doublet in our domain.

```python
# function to compute the velocity components of a doublet
def getVelocityDoublet(strength,xd,yd,X,Y):
    u = - strength/(2*pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    return u,v

# function to compute the stream-function of a doublet
def getStreamFunctionDoublet(strength,xd,yd,X,Y):
    psi = - strength/(2*pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    return psi
```

Once, the functions have been defined, we call them using the parameters of the doublet (its strength `kappa` and its location `xDoublet`, `yDoublet`).

```python
# computing the velocity components on the mesh grid
uDoublet,vDoublet = getVelocityDoublet(kappa,xDoublet,yDoublet,X,Y)

# computing the stream-function on the mesh grid
psiDoublet = getStreamFunctionDoublet(kappa,xDoublet,yDoublet,X,Y)
```

```python
# plotting
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uDoublet,vDoublet,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xDoublet,yDoublet,c='#CD2305',s=80,marker='o')
```

## Uniform flow past a doublet

A doublet alone does not give so much information about how it can be used to represent the streamlines around a shape. So lets superimpose our doublet on a uniform flow. We begin by defining the characteristics of the freestream (magnitude and angle of attack).

```python
Uinf = 1.0        # freestream speed
alpha = 0.0       # angle of attack (in degrees)
```

We recall that the Cartesian velocity components of a uniform flow are given by

$$u\left(x,y\right) = U_\infty \cos\alpha$$

$$v\left(x,y\right) = U_\infty \sin\alpha$$

and the stream-function by

$$\psi\left(x,y\right) = U_\infty \left(y\cos\alpha - x\sin\alpha\right)$$

```python
uFreestream = Uinf*cos(alpha*pi/180)*np.ones((N,N),dtype=float)
vFreestream = Uinf*sin(alpha*pi/180)*np.ones((N,N),dtype=float)

psiFreestream = Uinf*(cos(alpha*pi/180)*Y-sin(alpha*pi/180)*X)
```

The streamlines generated by a doublet in a uniform flow can be calculated using the principle of superposition (just a simple addition). Then, we can see that the *dividing streamline* is a cylinder which center is the location of the doublet. All the streamlines remaining outside the cylinder came from the uniform flow. Therefore, a doublet is a singularity that can be used to represent the streamlines around a  circular cylinder.

```python
# superimposition of the doublet on the freestream flow
u = uFreestream + uDoublet
v = vFreestream + vDoublet
psi = psiFreestream + psiDoublet

# plotting
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.contour(X,Y,psi,levels=[0.0],colors='#CD2305',linewidths=2,linestyles='solid')
plt.scatter(xDoublet,yDoublet,c='#CD2305',s=80,marker='o')
```

### Bernoulli's equation and the pressure coefficient

A very useful measurement of a flow around a body is the *coefficient of pressure* $C_p$. To evaluate the pressure coefficient, we apply *Bernoulli's equation* for incompressible flow:

$$p_\infty + \frac{1}{2}\rho V_\infty^2 = p + \frac{1}{2}\rho V^2$$

We define the pressure coefficient:

$$C_p = \frac{p-p_\infty}{\frac{1}{2}\rho V_\infty^2}$$

i.e.

$$C_p = 1 - \left(\frac{V}{V_\infty}\right)^2$$

In an incompressible flow, $C_p=1$ at a stagnation point.

```python
# compute the pressure coefficient
Cp = 1.0-(u**2+v**2)/Uinf**2

# plotting
size = 10
plt.figure(num=0,figsize=(1.1*size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
contf = plt.contourf(X,Y,Cp,levels=np.linspace(-2.0,1.0,100),extend='both')
cbar = plt.colorbar(contf)
cbar.set_label(r'$C_p$',fontsize=16)
cbar.set_ticks([-2.0,-1.0,0.0,1.0])
plt.scatter(xDoublet,yDoublet,c='#CD2305',s=80,marker='o')
plt.contour(X,Y,psi,\
            levels=[0.0],\
            colors='#CD2305',linewidths=2,linestyles='solid')
```
